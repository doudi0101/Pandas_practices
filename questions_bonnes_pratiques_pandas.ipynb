**La notation par crochets** fonctionnera toujours, alors que la **notation par points** a des limites :
- La notation par points ne fonctionne pas s'il y a des **espaces** dans le nom de la série.
- La notation par points ne fonctionne pas si la série porte le même nom qu'une méthode ou un attribut de **DataFrame** (comme 'head' ou 'shape').
- La notation par points ne peut pas être utilisée pour définir le nom d'une **nouvelle série** (voir ci-dessous).

```python
# Créez une nouvelle série 'Location' (utilisez la notation entre crochets pour définir le nom de la série).
ufo['Location'] = ufo.City + ', ' + ufo.State
ufo.head()
```

# Comment filtrer les lignes d'un DataFrame Pandas par valeur de colonne ?

```python
# Lire le jeu de données des films IMDb les mieux notés dans un DataFrame.
movies = pd.read_csv('http://bit.ly/imdbratings')
movies.head()
```

```python
# vérifier le nombre de lignes et de colonnes
movies.shape
```

**Objectif :** Filtrer les lignes du DataFrame pour n'afficher que les films dont la "durée" est d'au moins 200 minutes.

```python
# créer une liste dans laquelle chaque élément fait référence à une ligne de DataFrame : Vrai si le rang remplit la condition, Faux sinon.
booleans = []
for length in movies.duration:
    if length >= 200:
        booleans.append(True)
    else:
        booleans.append(False)
```

```python
# confirmer que la liste a la même longueur que le DataFrame
len(booleans)
```

```python
# examiner les cinq premiers éléments de la liste
booleans[0:5]
```

```python
# convertir la liste en Series
is_long = pd.Series(booleans)
is_long.head()
```

```python
# utiliser la notation entre crochets avec le booléen Series pour indiquer au DataFrame les lignes à afficher.
movies[is_long]
```

```python
# simplifier les étapes ci-dessus : il n'est pas nécessaire d'écrire une boucle for pour créer 'is_long' puisque pandas diffusera la comparaison.
is_long = movies.duration >= 200
movies[is_long]

# ou de manière équivalente, écrivez-le en une ligne (pas besoin de créer l'objet 'is_long')
movies[movies.duration >= 200]
```

```python
# sélectionner la série 'genre' dans le DataFrame filtré
movies[movies.duration >= 200].genre

# ou de manière équivalente, utiliser la méthode 'loc'.
movies.loc[movies.duration >= 200, 'genre']
```

Documentation pour [**`loc`**](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.loc.html)

# Comment appliquer plusieurs critères de filtrage à un DataFrame Pandas ? 

```python
movies = pd.read_csv('http://bit.ly/imdbratings')
movies.head()
```

```python
# Filtrez le DataFrame pour afficher uniquement les films dont la durée est d'au moins 200 minutes.
movies[movies.duration >= 200]
```

Comprendre les **opérateurs logiques:**

- **`and`** : Vrai seulement si **les deux côtés** de l'opérateur sont Vrais
- **`or`** : Vrai si **l'un des deux côtés** de l'opérateur est vrai.

```python
# démonstration de l'opérateur "and".
print(True and True)
print(True and False)
print(False and False)
```

```python
# démonstration de l'opérateur 'or'.
print(True or True)
print(True or False)
print(False or False)
```

Règles pour spécifier **des critères de filtre multiples** dans pandas :

- utiliser **`&`** au lieu de **`and`**.
- utilisez **`|`** au lieu de **`or`**.
- ajouter des **parenthèses** autour de chaque condition pour spécifier l'ordre d'évaluation

**Objectif:** Filtrer davantage le DataFrame des films longs (duartion >= 200) pour n'afficher que les films dont le 'genre' est également 'Drama'.

```python
# utiliser l'opérateur '&' pour spécifier que les deux conditions sont requises.
movies[(movies.duration >=200) & (movies.genre == 'Drama')]
```

```python
# INCORRECT : l'utilisation de l'opérateur '|' aurait montré des films qui sont soit longs, soit dramatiques (ou les deux).
movies[(movies.duration >=200) | (movies.genre == 'Drama')].head()
```

**But:** Filtrer le DataFrame original pour montrer les films avec un 'genre' de 'Crime' ou 'Drama' ou 'Action'.

```python
# Utilisez l'opérateur '|' pour spécifier qu'une ligne peut correspondre à l'un des trois critères.
movies[(movies.genre == 'Crime') | (movies.genre == 'Drama') | (movies.genre == 'Action')].head(10)
```

```python
# ou de manière équivalente, utiliser la méthode 'isin'.
movies[movies.genre.isin(['Crime', 'Drama', 'Action'])].head(10)
```

Documentation for [**`isin`**](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.Series.isin.html)

Lorsque vous **référer à des lignes ou à des colonnes** avec le paramètre axis :

- **axis 1** fait référence aux colonnes

Lorsque vous effectuez une **opération mathématique** avec le paramètre d'axe :

- **axis 0** signifie que l'opération doit se déplacer vers le bas sur l'axe des lignes.
- **axis 1** signifie que l'opération doit "se déplacer sur" l'axe des colonnes.

```python
# 'index' est un alias pour l'axe 0
drinks.mean(axis='index')
```

```python
# 'columns' est un alias pour l'axe 1
drinks.mean(axis='columns').head()
```

#  Quand dois-je utiliser un "groupby" dans pandas ?

# Pourquoi les tranches de DataFrame sont-elles inclusives lorsqu'on utilise **`.loc`**, mais exclusives lorsqu'on utilise **`.iloc`** ?

```python
# Le découpage basé sur les étiquettes comprend le début et la fin du découpage.
ufo.loc[0:4, :]
```

```python
# Le découpage basé sur la position inclut le début et exclut la fin.
ufo.iloc[0:4, :]
```

Documentation for [**`loc`**](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.loc.html) and [**`iloc`**](http://pandas.pydata.org/pandas-docs/stable/generated/pandas.DataFrame.iloc.html)

```python
# 'iloc' suit simplement la convention de découpage de NumPy...
ufo.values[0:4, :]
```

```python
# ...et NumPy suit simplement la convention de découpage de Python
'python'[0:4]
```

Documentation sur [le découpage en python](https://python-reference.readthedocs.io/en/latest/docs/brackets/slicing.html)
